# Aufgabe 4.2

Erzeugen Siemit `norm.rvs()` $m = 500$ Stichproben aus einer Standardnormalverteilung mit Umfang $n = 5$. Speichern Sie die Stichproben als (n × m)− Matrix ab.

```python
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import norm
import numpy as np

matplotlib.rcParams['figure.figsize'] = [16, 8]

n = 5
m = 500
ran = np.array(norm.rvs(size=n*m))
sim = ran.reshape((n,m))
```

**a) Stellen Sie die Stichproben als Runs graphisch dar, und staunen Sie über die zahlreichen möglichen Verläufe.**

```python
plt.plot(sim)
plt.show()
```

müssten. Überprüfen Sie dies graphisch mit Hilfe von Histogrammen und der
theoretischen Wahrscheinlichkeitsdichtekurve

```python
plt.hist(sim.T, bins=20, density=True, edgecolor="black",
facecolor="white")
x = np.linspace(-4, 4, num=100)
y = norm.pdf(x)
plt.plot(x,y)
plt.title("Histogramm sim")
plt.show()
sim_mean = sim.mean(axis=0)
plt.hist(sim_mean, density=True, edgecolor="black",
facecolor="white")
x = np.linspace(-4, 4, num=100)
y = norm.pdf(x, loc=0, scale=1/np.sqrt(n))
plt.plot(x,y)
plt.title("Histogramm sim_mean")
plt.show()
```

# Aufgabe 4.5

Wir betrachten einen Datensatz, bei welchem zwei Methoden zur Bestimmung der latenten Schmelzwärme von Eis verglichen werden. Wiederholte Messungen der freigesetzten Wärme beim Übergang von Eis bei $−0.7°C$ zu Wasser bei $0°C$ ergaben die folgenden Werte (in cal/g):

```python
import numpy as np
from math import sqrt

ma = np.array([79.98, 80.04, 80.02, 80.04, 80.03, 80.03, 80.04, 79.97, 80.05, 80.03, 80.02, 80.00, 80.02])
mb = np.array([80.02, 79.94, 79.98, 79.97, 79.97, 80.03, 79.95, 79.97])
```

**a) Berechnen Sie die arithmetischen Mittelwerte der beiden Methoden und geben Sie für jede Methode den absoluten Fehler an.**

Standardfehler:

$$ \overline{x}_n \pm s_{\overline{x}_n} = \overline{x}_n \pm \frac{s_x}{\sqrt{n}} $$
relativer Fehler:

$$ \overline{x}_n \pm \frac{s_{\overline{x}_n}}{\overline{x}_n} $$
$ s_x $: empirische Standardabweichung der Messreihe:

$$ s_x = \sqrt{\frac{1}{n-1} \sum_{i=1}^n(x_i - \overline{x})^2} $$

```python
mean_a = ma.mean()
mean_b = mb.mean()
print('mean a = {:.3f}, mean b = {:.3f}'.format(mean_a, mean_b))

sx_a = sqrt((1 / (len(ma)-1)) * np.sum((ma - mean_a) ** 2))
sx_b = sqrt((1 / (len(mb)-1)) * np.sum((mb - mean_b) ** 2))
print('sxa={:.3f}, sxb={:.3f}'.format(sx_a, sx_b))

abs_err_a = sx_a / sqrt(len(ma))
abs_err_b = sx_b / sqrt(len(mb))
print('manually: abs_err_a={:.5f}, abs_err_b={:.5f}'.format(abs_err_a, abs_err_b))

# using NumPy: ma.std() / np.sqrt(len(ma))
abs_err_a = ma.std(ddof=1) / sqrt(len(ma))
abs_err_b = mb.std(ddof=1) / sqrt(len(mb))
print('NumPy: abs_err_a={:.5f}, abs_err_b={:.5f}'.format(abs_err_a, abs_err_b))
```


Messung A: $ (80.02 \pm 0.01)cal/g $

Messung B: $ (79.98 \pm 0.01)cal/g $

**b) Wie lauten die relativen Fehler?**

```python
rel_err_a = abs_err_a / ma.mean()
rel_err_b = abs_err_b / mb.mean()

print('rel_err_a={:.5f}, rel_err_b={:.5f}'.format(rel_err_a, rel_err_b))
```

Messung A: $ 80.02\ cal/g \pm 0.008 \% $

Messung B: $ 79.98\ cal/g \pm 0.0014 \% $

# Aufgabe 4.6

Die Auswertung eines Integrals

$$ \int_{a}^{b} f(x) \mathrm{d} x $$

kann sehr oft nicht analytisch erfolgen. Der gebräuchlichste Ansatz in diesem Fall
besteht darin, das Integral numerisch zu berechnen. Dazu existieren verschiedene
Computerprogramme. Eine andere geläufige Methode, um ein solches Integral zu
berechnen, ist die sogenannte **Monte Carlo Methode**. Man generiert dabei uniform
verteilte Zufallsvariablen auf dem Intervall $[a, b]$, d. h., $X_1, X_2, ..., X_n i.i.d. ∼ Uniform([a, b])$ und berechnet

$$
\frac{b-a}{n} \sum_{i=1}^{n} f\left(X_{i}\right)
$$

Dass dieser Ausdruck in etwa $\int_{a}^{b} f(x) \mathrm{d} x$ ist, möchten wir im Folgenden verstehen.
Aufgrund des *Gesetzes der grossen Zahlen* gilt für grosse $n$

$$
(b-a) \cdot \frac{1}{n} \sum_{i=1}^{n} f\left(X_{i}\right) \approx(b-a) \cdot \mathrm{E}[f(X)]
$$

wobei $\frac{1}{b-a}$ die Wahrscheinlichkeitsdichte der Wahrscheinlichkeitsverteilung $Uniform([a, b])$
ist. Somit gilt

$$
\frac{b-a}{n} \sum_{i=1}^{n} f\left(X_{i}\right) \approx(b-a) \cdot \mathrm{E}[f(X)]=(b-a) \cdot \int_{a}^{b} f(x) \frac{1}{b-a} \mathrm{d} x=\int_{a}^{b} f(x) \mathrm{d} x
$$

Berechnen Sie folgendes Integral

$$
I(f)=\frac{1}{\sqrt{2 \pi}} \int_{0}^{1} \mathrm{e}^{-x^{2} / 2} \mathrm{d} x
$$

Berechnen Sie das Integral, indem Sie 1000 uniform über das Intervall $[0, 1]$ verteilte
Zahlen $X_1, ..., X_1000$ mit der Python-Funktion `uniform.rvs(...)` generieren.
Berechnen Sie den genauen numerischen Wert des Integrals mit der Python-Funktion
`norm.cdf(...)`.

```python
import numpy as np
from scipy.stats import uniform, norm
from math import sqrt, exp, pi

def f(x):
    return 1 / sqrt(2 * pi) * np.exp(np.negative(x**2)/2)

xs = uniform.rvs(loc=0, scale=1, size=1000)
ys = f(xs)

area = (1-0)/len(xs) * ys.sum()
print('estimated: {:.3f}'.format(area))

area = norm.cdf(1) - norm.cdf(0)
print('calculated: {:.3f}'.format(area))
```
